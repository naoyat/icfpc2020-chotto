```python
from scipy.io import wavfile
rate, data = wavfile.read('radio-transmission-recording.wav')
```

```python
print("rate=", rate) # 44100
from scipy import signal
f, t, sxx = signal.spectrogram(data, fs=rate, nperseg=256, window='hann') # nperseg=周波数の分割数=1024 - 44100を1024で割るから44Hz刻みぐらい? ;; 1024だと800ぐらいまで、512にすると1600まで出る
print(f.shape) # (513,) Array of sample frequencies.
print(t.shape) # (3041,) Array of segment times.
print(sxx.shape) # (513, 3041) Spectrogram of x. By default, the last axis of Sxx corresponds to the segment times.


import matplotlib.pyplot as plt
%matplotlib inline  


```

```python
seg, length = sxx.shape
f_unit = f[1] - f[0]

def find_peak_indices(sxx):
    # find_peak_indices
    av1 = np.average(sxx, axis=1)
    av = np.average(av1)
    
    indices = av1.argsort()[-2:]
    indices.sort()
    lo, hi = indices
    return lo, hi

def onefreq(sxx, lo, hi):
    cutoff = np.average(sxx[hi+1:, :])
    # print("cutoff = ", cutoff)
    
    one = np.zeros(length)
    two = np.zeros(length)
    total = 0
    for i in range(length):
        freq = 0
        denom = 0
        for ix in range(lo-4, hi+5):
            f = ix * f_unit
            v = max(0, sxx[ix,i] - cutoff)
            freq += f * v
            denom += v
        if denom == 0:
            one[i] = 0
            two[i] = 0
        else:
            one[i] = freq / denom
            two[i] = denom
            total += denom
    
    avg = total / length
    # print("avg=", avg)

    threshold = avg / 5
    for i in range(length):
        if one[i] > 0 and two[i] < threshold:
            one[i] = 0
    return one


lo, hi = find_peak_indices(sxx)
mid = (f[lo]+ f[hi])/2
# print("lo=%d=%g, hi=%d=%g, mid=%g" % (lo, f[lo], hi, f[hi], mid))

of = onefreq(sxx,lo,hi)
# print(of[:600])

def onetwo(of):
    out = np.zeros(length, dtype=int)
    m = 50
    for i in range(length):
        if 500-m < of[i] < 500+m:
            out[i] = 1
        elif 600-m < of[i] < 600+m:
            out[i] = 2
        else:
            out[i] = 0
    return out

    
of12 = onetwo(of)
# print(of12[:600])


def rle(of):
    res = []
    def flush(last, cnt):
        if last > 0:
            c = round(cnt/9.8)
            return [last-1] * c
        else:
            return []
            # res.append((last-1, c))
        
    last = -1
    cnt = 0
    for i in range(length):
        if of[i] == last:
            cnt += 1
        else:
            if cnt > 0:
                res += flush(last, cnt)
            last = of[i]
            cnt = 1
    if cnt > 0:
        res += flush(last, cnt)

    return res

d = rle(of12)
# print("|d| =", len(d))
# print(d)

def show(data, width, remove_frame=True):
    def conv(row):
        return ''.join("#" if c==1 else '.' for c in row)

    L = len(data)
    height = (L + width - 1) // width
    rows = []
    if width * height > L:
        data += [0] * (width * height - L)
    for c in range(height):
        row = data[c*width : (c+1)*width]
        if remove_frame:
            if c == 0 or c == height-1:
                continue
            row = row[1:-1]
            assert len(row) == width-2
            
        rows.append( conv(row) )
        # print(conv(row))
    return rows


def find_width(data):
    L = len(data)
    for i in range(L):
        if data[i] == 0:
            return i-1
    return -1


width = find_width(d)
print('width =', width)

rows = show(d, width)
for row in rows:
    print(row)


# # print(sum(c for v,c in r))


#print(pickup(sxx, lo, hi))

# print('----')
# print(f[10:20])

# print(sxx[10:20,:])

import numpy as np

# a = np.sum(sxx,)

#plt.figure(figsize=(50, 10))
#window = 3041
#window = 400 # 50で約1って何？ 音源は約1分で、timeframeは3041、ってことは50フレームで1秒。なんでそうなる？
#pre = 130# 140
# plt.pcolormesh(t[pre:pre+window], f[10:20], sxx[10:20, pre:pre+window], shading='gouraud')
# plt.show()
```
